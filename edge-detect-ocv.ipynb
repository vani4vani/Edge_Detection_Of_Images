```python
# import opencv for computer vision
import cv2
# import os for file paths
import os
#import matplotlib for visualisation
import matplotlib.pyplot as plt
```

 ## Load Image

```python
#define the file path to the image
IMG_PATH = os.path.join('images', 'tiger.jpg')
```

```python
#read in image
img = cv2.imread(IMG_PATH)
```

```python
# resize image
resized_img = cv2.resize(img, (int(img.shape[1]/2), int(img.shape[0]/2)))
#view image using opencv
cv2.imshow('Frame View', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Edge Detection

```python
#apply canny
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (5,5), 0)
canny = cv2.Canny(blur, threshold1=180, threshold2=200)
```

```python

# resize image
resized_img = cv2.resize(img, (int(img.shape[1]/2), int(img.shape[0]/2)))
#view image using opencv
cv2.imshow('Frame View', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Export Edges

```python
cv2.imwrite('edges.jpg', canny)
```
